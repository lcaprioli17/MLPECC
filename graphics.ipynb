```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('/Users/luanabussu/Documents/GitHub/ProgettoDM/dataset/pp_dataset/k2_filled.csv', on_bad_lines='skip')

data_corr = data.corr()
f, ax = plt.subplots(figsize = (40, 34))
sns.heatmap(data_corr, vmax = 1, square = True, annot=True)
```

```python
sorted_pairs = data_corr.unstack().sort_values (kind="quicksort")
filtered_pairs = sorted_pairs.drop_duplicates()
filtered_pairs = filtered_pairs[abs(filtered_pairs) > 0.95]
print(filtered_pairs)
```
